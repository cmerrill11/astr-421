```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
hist1 = pd.read_csv('1MsunData/history.data')
hist25 = pd.read_csv('25MsunData/history.data')
```

```python
fig, ax = plt.subplots(1,1, figsize = (7,7))
# Special Timesteps:
step = [95, 147, 290, 5903, 5632]

for a in hist1['model_number']:
    if a in step:
        d = hist1[hist1['model_number'] == a]
        ax.scatter(d['log_Teff'], d['log_L'], label = a, marker = '^')
    else: 
        cmap = ax.scatter(hist1['log_Teff'], hist1['log_L'], 
            c=np.log10(hist1['star_age']), cmap=plt.cm.Spectral, marker='o', alpha=0.5)


plt.gca().invert_xaxis()
cb = plt.colorbar(cmap)
plt.legend()
cb.set_label('log Age (yrs)')
plt.xlabel('log Teff')
plt.ylabel('log L')
```
